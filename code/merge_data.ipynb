```python
# Merge model specs with results on MMLU
import json

import pandas as pd

import utils.globals as uglobals
```

```python
df = pd.read_excel(f'{uglobals.DATA_DIR}/LLM_specs.xlsx')
print(df)
```

```python
# exp_name = 'aggregated_5shot'
exp_name = 'aggregated_5shot_grouped'
with open(f'{uglobals.AGGREGATED_OUTPUT_DIR}/{exp_name}.json', 'r') as f:
    data = json.load(f)

for model, vals in data.items():
    if '_' in model:
        model_name = model.split('_')[0] + '/' + '_'.join(model.split('_')[1:])
    else:
        model_name = model

    for idx in range(len(df)):
        if df['Model name'][idx] == model_name:
            model_id = idx

    vals['n_params'] = int(df['Parameters (M)'][model_id])
    vals['multilingual'] = int(df['Multilingual'][model_id])
    vals['code'] = int(df['Code'][model_id])
    vals['distillation'] = int(df['Distillation'][model_id])
    vals['rlhf'] = int(df['RLHF'][model_id])
    vals['gpt2'] = int(df['GPT2'][model_id])
    vals['gpt3'] = int(df['GPT3'][model_id])
    vals['gpt3.5'] = int(df['GPT3.5'][model_id])
    data[model] = vals

with open(f'{uglobals.AGGREGATED_OUTPUT_DIR}/{exp_name}_spec_full.json', 'w') as f:
    json.dump(data, f)    
```
